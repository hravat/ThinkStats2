```python
from collections import Counter
C = Counter(a)
total = float(sum(C.values()))
for key in C: 
    print(str(key)+':- '+str(C[key]/total))
```

Thus the counter is a convient way to count frequencies

We will first replace birthord to 0 for all not first pregnancies.
